## Q1. What is a parameter?

**Answer:**

A parameter is a variable used by a machine learning model to make predictions. These values are learned during training. Example: weights in linear regression.

## Q2. What is correlation?

**Answer:**

Correlation measures the relationship between two variables and shows how they move together.

## Q3. What does negative correlation mean?

**Answer:**

Negative correlation means that when one variable increases, the other variable decreases.

## Q4. Define Machine Learning. What are the main components in Machine Learning?

**Answer:**

Machine Learning is a branch of Artificial Intelligence where systems learn patterns from data and make predictions or decisions without being explicitly programmed. Main components: Data, Features, Model, Loss Function, Optimizer, and Evaluation.

## Q5. How does loss value help in determining whether the model is good or not?

**Answer:**

Loss value measures the error between predicted and actual values. Lower loss means the model is performing better.

## Q6. What are continuous and categorical variables?

**Answer:**

Continuous variables can take any numerical value (e.g., height, weight). Categorical variables represent categories or labels (e.g., gender, color).

## Q7. How do we handle categorical variables in Machine Learning? What are the common techniques?

**Answer:**

Categorical variables are converted into numerical form using techniques like Label Encoding and One-Hot Encoding.

## Q8. What do you mean by training and testing a dataset?

**Answer:**

Training data is used to train the model, while testing data is used to evaluate the model's performance.

## Q9. What is sklearn.preprocessing?

**Answer:**

sklearn.preprocessing is a module in Scikit-learn used for scaling, normalization, encoding, and transforming data.

## Q10. What is a Test set?

**Answer:**

A test set is a portion of the dataset used to evaluate the trained model on unseen data.

## Q11. How do we split data for model fitting (training and testing) in Python?

**Answer:**

We use train_test_split() from sklearn.model_selection to split data into training and testing sets.

## Q12. How do you approach a Machine Learning problem?

**Answer:**

Steps: Understand problem, collect data, perform EDA, preprocess data, train model, evaluate model, and improve performance.

## Q13. Why do we have to perform EDA before fitting a model to the data?

**Answer:**

EDA helps understand data patterns, detect missing values, outliers, and relationships before training the model.

## Q14. How can you find correlation between variables in Python?

**Answer:**

We can use the corr() function in Pandas to calculate correlation between variables.

## Q15. What is causation? Explain difference between correlation and causation with an example.

**Answer:**

Causation means one variable directly affects another. Correlation only shows a relationship. Example: Ice cream sales and drowning cases increase together in summer. They are correlated, but ice cream does not cause drowning.

## Q16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

**Answer:**

An optimizer updates model parameters to reduce loss. Common optimizers are Gradient Descent, SGD, Adam, and RMSProp. Example: Adam optimizer is widely used in deep learning.

## Q17. What is sklearn.linear_model?

**Answer:**

sklearn.linear_model is a Scikit-learn module used for linear models such as Linear Regression and Logistic Regression.

## Q18. What does model.fit() do? What arguments must be given?

**Answer:**

model.fit() trains the model using training data. Arguments usually include features (X) and target values (y).

## Q19. What does model.predict() do? What arguments must be given?

**Answer:**

model.predict() makes predictions using the trained model. It takes feature data (X) as input.

## Q20. What is feature scaling? How does it help in Machine Learning?

**Answer:**

Feature scaling standardizes data values so all features are on a similar scale, improving model performance and training speed.

## Q21. How do we perform scaling in Python?

**Answer:**

Scaling can be done using StandardScaler or MinMaxScaler from sklearn.preprocessing.

## Q22. Explain data encoding?

**Answer:**

Data encoding is the process of converting categorical data into numerical format so that machine learning algorithms can understand it.
