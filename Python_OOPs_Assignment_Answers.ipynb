# Python OOPs Assignment Answers


## Q1. What is Object-Oriented Programming (OOP)?

**Answer:** Object-Oriented Programming (OOP) is a programming paradigm based on objects and classes. It helps organize code using concepts like inheritance, encapsulation, abstraction, and polymorphism.

## Q2. What is a class in OOP?

**Answer:** A class is a blueprint for creating objects. It defines attributes and methods.

## Q3. What is an object in OOP?

**Answer:** An object is an instance of a class that contains data and methods.

## Q4. What is the difference between abstraction and encapsulation?

**Answer:** Abstraction hides implementation details and shows only essential features, while encapsulation hides data using private variables and methods.

## Q5. What are dunder methods in Python?

**Answer:** Dunder methods are special methods with double underscores like __init__ and __str__. They define built-in behaviors of objects.

## Q6. Explain the concept of inheritance in OOP.

**Answer:** Inheritance allows one class to acquire properties and methods of another class.

## Q7. What is polymorphism in OOP?

**Answer:** Polymorphism allows the same method name to behave differently for different objects.

## Q8. How is encapsulation achieved in Python?

**Answer:** Encapsulation is achieved using private variables and methods with double underscores (__).

## Q9. What is a constructor in Python?

**Answer:** A constructor is a special method __init__() used to initialize object attributes.

## Q10. What are class and static methods in Python?

**Answer:** Class methods use @classmethod and work with class variables. Static methods use @staticmethod and do not depend on class or object.

```python

"""Q1. Create a parent class Animal with a method speak() that prints a generic message. 
Create a child class Dog that overrides the speak() method to print "Bark!"."""
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Bark!")

dog = Dog()
dog.speak()


""" Q2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle 
from it and implement the area() method in both."""
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

print(Circle(5).area())
print(Rectangle(4, 6).area())


""" 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute."""
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    pass

class ElectricCar(Car):
    def __init__(self, type, battery):
        super().__init__(type)
        self.battery = battery

e = ElectricCar("SUV", "100kWh")
print(e.type, e.battery)


""" 4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
Sparrow and Penguin that override the fly() method.
"""
class Bird:
    def fly(self):
        print("Bird flies")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

for bird in [Sparrow(), Penguin()]:
    bird.fly()


"""5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance."""
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def check_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
print(acc.check_balance())


""" 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar
and Piano that implement their own version of play()."""
class Instrument:
    def play(self):
        print("Instrument plays")

class Guitar(Instrument):
    def play(self):
        print("Guitar playing")

class Piano(Instrument):
    def play(self):
        print("Piano playing")

for i in [Guitar(), Piano()]:
    i.play()


""" 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static
method subtract_numbers() to subtract two numbers."""
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

print(MathOperations.add_numbers(10, 5))
print(MathOperations.subtract_numbers(10, 5))


""" 8. Implement a class Person with a class method to count the total number of persons created."""
class Person:
    count = 0

    def __init__(self):
        Person.count += 1

    @classmethod
    def total_persons(cls):
        return cls.count

p1 = Person()
p2 = Person()
print(Person.total_persons())


""" 9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the
fraction as "numerator/denominator"."""
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

f = Fraction(1, 2)
print(f)


# 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)

""" 11. Create a class Person with attributes name and age. Add a method greet() that prints "Hello, my name is
{name} and I am {age} years old."""

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")


# Creating object
person1 = Person("Bhupendra", 27)

# Calling method
person1.greet()

""" 12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
the average of the grades."""

class Student:
    
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

student1 = Student("Rahul", [85, 90, 78, 92])

print("Student Name:", student1.name)
print("Average Grade:", student1.average_grade())


""" 13. Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the
area"""

class Rectangle:

    
    def set_dimensions(self, length, width):
        self.length = length
        self.width = width

    
    def area(self):
        return self.length * self.width



rect = Rectangle()

rect.set_dimensions(10, 5)

print("Length:", rect.length)
print("Width:", rect.width)
print("Area of Rectangle:", rect.area())


""" 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary."""

# Base class Employee
class Employee:

    # Constructor
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    # Method to calculate salary
    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate


# Derived class Manager
class Manager(Employee):

    # Constructor
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    # Method overriding
    def calculate_salary(self):
        return (self.hours_worked * self.hourly_rate) + self.bonus


# Create Employee object
emp = Employee("Rahul", 40, 500)

# Create Manager object
mgr = Manager("Amit", 40, 800, 5000)

# Display salaries
print("Employee Salary:", emp.calculate_salary())
print("Manager Salary:", mgr.calculate_salary())


""" 15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that
calculates the total price of the product."""

class Product:

    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity


product1 = Product("Laptop", 50000, 2)

print("Product Name:", product1.name)
print("Total Price:", product1.total_price())

""" 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method."""

from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass


# Derived class Cow
class Cow(Animal):

    def sound(self):
        print("Cow says Moo")


# Derived class Sheep
class Sheep(Animal):

    def sound(self):
        print("Sheep says Baa")


# Create objects
c = Cow()
s = Sheep()

# Call methods
c.sound()
s.sound()

""" 17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that
returns a formatted string with the book's details."""

# Create a class Book
class Book:

    # Constructor
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    # Method to return book information
    def get_book_info(self):
        return f"Title: {self.title}, Author: {self.author}, Year: {self.year_published}"


# Create object
book1 = Book("Python Basics", "John Smith", 2024)

# Display book info
print(book1.get_book_info())

""" 18. Create a class House with attributes address and price. Create a derived class Mansion that adds an
attribute number_of_rooms."""

# Base class House
class House:

    # Constructor
    def __init__(self, address, price):
        self.address = address
        self.price = price


# Derived class Mansion
class Mansion(House):

    # Constructor
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms


# Create object
m1 = Mansion("Mumbai", 50000000, 10)

# Display details
print("Address:", m1.address)
print("Price:", m1.price)
print("Number of Rooms:", m1.number_of_rooms)
```
